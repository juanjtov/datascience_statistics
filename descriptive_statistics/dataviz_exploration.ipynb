## Scattered Plot


```
import pandas as pd
import seaborn as sns
```

```
#load the data set
iris = sns.load_dataset('iris')
iris.head(10)
```

### Seaborn scatter plot
Source: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```
sns.scatterplot(data=iris, x='sepal_length', y='petal_length')
```

- Seggregate a categorical var

```
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

### Seaborn Join Plot

```
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

```
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

### Finding new relationships among features

```
type(iris), iris.columns
```

```
sns.jointplot(data=iris, x='sepal_length', y='sepal_width')
```

```
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
```

```
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

- So far this graph has shown the strongest relationship, you can easily split the different species.
- Another weak conclusion is that there is sort of a linear beahaviour between the petal length and width.

```
sns.jointplot(data=iris,x='petal_length', y='petal_width', hue='species')
```

- No, we can see that all the species try to follow a normal distribution, which means that the mayoritiy of the values are close to the mean. 
  Especially the setosa specie which has a strong normal distribution behaviour.

### Box Plot

```
sns.boxenplot(x='species', y='sepal_length', data =iris)
```

### Bar Plot

```
pl_mean = iris['petal_length'].mean()
pw_mean = iris['petal_width'].mean()

print(f'Mean of the Petal length: {pl_mean}')
print(f'Mean of the Petal Width: {pw_mean}')
```

- It's more significant if we do it by species

```
iris.groupby('species').mean()
```

```
# Mean by specie
iris.groupby('species')['petal_length'].mean()
```

```
sns.barplot(data=iris, x='species', y='petal_length')
```

```
sns.barplot(data=iris, x='species', y='petal_width')
```

```
sns.histplot(iris['petal_length'])
```
